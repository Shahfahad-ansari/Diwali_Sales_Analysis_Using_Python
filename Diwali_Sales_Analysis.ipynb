```python
# drop unrelated/blank column
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
# drop null values
df.dropna(inplace=True)
```

# Exploratory Data Analysis
**Gender**

```python
# plotting a bar chart for gender and it's count

ax = sns.countplot(x = 'Gender', data= df, hue= 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# plotting a bar chart for gender vs total amount

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender', y= 'Amount', data = sales_gen, hue= 'Gender')
```

*From above graphs we can see that most of the buyers are females and even the purchasing power of the females are greater than males*

**Age**

```python
# Age group and it's count

ax = sns.countplot(x = 'Age Group', data= df, hue= 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Age group vs total amount

sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group', y= 'Amount', data = sales_age, hue= 'Age Group')
```

*From above graphs we can see that most of the buyers are of age group between 26-35 yrs female*

**State**

```python
# total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

plt.figure(figsize=(16, 5))
sns.barplot(x = 'State', y= 'Orders', data = sales_state, hue= 'State')
```

```python
# total amount/sales from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

plt.figure(figsize=(16, 5))
sns.barplot(x = 'State', y= 'Amount', data = sales_state, hue= 'State')
```

*From above graphs we can see most of the orders and total sales/amount are from Uttar Pradesh, Maharashtra, Karnataka respectively*

**Marital Status**

```python
# Marital Status and it's count

ax = sns.countplot(x = 'Marital_Status', data = df, hue= 'Marital_Status')


for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Martial status vs total amount

sales_MaritalStatus = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

plt.figure(figsize=(6, 5))
sns.barplot(x = 'Marital_Status', y= 'Amount', data = sales_MaritalStatus, hue='Gender')
```

*From above graphs we can see that most of the buyers are unmarried (women) and they have high purchasing power*

**Occupation**

```python
# Occupation and it's count

plt.figure(figsize=(20, 5))
ax = sns.countplot(x = 'Occupation', data = df, hue= 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Occupation vs total amount

sales_Occupation = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

plt.figure(figsize=(20, 5))
sns.barplot(x = 'Occupation', y= 'Amount', data = sales_Occupation, hue='Occupation')
```

*From above graphs we can see that most of the buyers are working in IT, Healhcare and Aviation sector*

**Product Category**

```python
# Product category and it's count

plt.figure(figsize=(20, 5))
ax = sns.countplot(x='Product_Category', data=df, hue='Product_Category')

# Rotate the x-axis labels
plt.xticks(rotation=45)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Top 10 sold product category

sales_ProductCategory = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

plt.figure(figsize=(20, 5))
sns.barplot(x = 'Product_Category', y= 'Amount', data = sales_ProductCategory, hue= 'Product_Category')
```

*From above graphs we can see that most of the sold products are from Food, Clothing, and Electronics Category*

```python
# Top 10 sold product category

sales_PoductID = df.groupby(['Product_ID'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

plt.figure(figsize=(20, 5))
sns.barplot(x = 'Product_ID', y= 'Amount', data = sales_PoductID, hue= 'Product_ID')
```

**Conclusion:**

*Unmarried women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category*
